```python
df_all_returns = pd.DataFrame()

for ticker in ticker_list:
    dd['Ticker'] = ticker
    daily_returns = dd['Adj Close'].pct_change()
    yearly_returns = dd['Adj Close'].resample('Y').bfill().pct_change()
    yearly_returns = yearly_returns.to_frame()
    yearly_returns['Ticker'] = ticker

    df_all_returns = df_all_returns.append(yearly_returns)
```

```python
df_all_returns['Year'] = pd.DatetimeIndex(df_all_returns.index).year

df_all_returns.rename(columns={'Adj Close' : 'Annual Return'}, inplace=True)
```
